<h3>▪️Data Pre-processing </h3>

#### 1.Checking for duplicates

```python
duplicate_count = df.duplicated().sum()
print("Total duplicate rows:", duplicate_count)
```

#### 2.Removing duplicates

```python
df = df.drop_duplicates()
```

```python
duplicate_count = df.duplicated().sum()
duplicate_count
```

#### 3. Handling missing values

```python
print("Missing values before:\n")
df.isnull().sum()
```

#### 4. Removing dupicates in Amount column by removing column 

```python
df.dropna(subset=['Amount'], inplace=True)
```

```python
df.isnull().sum()
```

#### 5.Drop Unwanted columns

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.columns
```

#### 6.Replace shorthand values in gender column

```python
df["Gender"]=df["Gender"].replace({
    "M":"Male",
    "F":"Female"
})
df["Gender"].unique()
```

#### 7.Creating derived columns

```python
#create a column Marital_StatusNew
def matr_status(Marital_Status):
    if Marital_Status ==0:
        return 'UnMarried'
    elif Marital_Status ==1:
        return 'Married'
    else:
        return 'Other'
df['Marital_StausNew']=df['Marital_Status'].apply(matr_status)
```

```python
df.columns
```

```python
df['Marital_StausNew'].unique()
```

```python
# Deleted Marital_Status column as new column is derrived
```

```python
df.drop(['Marital_Status'], axis=1, inplace=True)
```

```python
df.columns
```

<div style="border: 2px solid #2e5a88;font-size: 14px;">

#### 11. Correlation Heatmap
